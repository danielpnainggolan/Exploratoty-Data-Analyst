**Exploratory Data Analysis**

**Keterangan:**

Langkah pertama adalah saya mengelompokkan data berdasrakan parameter tertentu.

**1. Pengelompokan data customer yang melakuakn request untuk setiap tahun** 

```python
g = pd.DatetimeIndex(df1['Tanggal Request']).year
f = df1.groupby(g)['Type'].count().reset_index().plot(x="Tanggal Request", y="Type", kind="bar", rot=3, fontsize=4)
f
```

**Keterangan:**

Dari data yang diperoleh, dapat dilihat bahwa jumlah customer yang melakukan request untuk setiap tahunnya tidak beraturan.Jumlah request paling tinggi terjadi pada tahun 2014 dan 2013. Namun berdasarkan data tersebut, setelah melewati tahun 2014, tingkat custumer yang request semakin menurun.
